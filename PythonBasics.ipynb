```python
"""
Python Comprehensive Guide for Machine Learning

This script covers:
1. Python Basics (Variables, Loops, Functions, Conditionals, Exception Handling)
2. Data Structures (Lists, Dictionaries, Tuples, Sets, Queues, Stacks)
3. File Handling (Reading/Writing files, JSON Handling)
4. NumPy Basics (Arrays, Operations, Indexing, Reshaping, Statistical Functions)
5. Pandas Basics (DataFrames, Series, Missing Data Handling, CSV & Excel Handling)
6. Data Visualization (Matplotlib, Seaborn, Customizing Plots)
7. Mathematics for ML (Linear Algebra, Calculus, Probability, Statistics)
8. Working with Datasets (Titanic Dataset Example)
"""

# ---------------- Python Basics ----------------

# Variables
x = 10
y = 20.5
name = "Machine Learning"
print(f"x: {x}, y: {y}, name: {name}")

# Conditionals
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")

# Loops
for i in range(5):
    print(f"Iteration: {i}")

# Functions
def greet(name):
    return f"Hello, {name}!"

print(greet("Python Learner"))

# Exception Handling
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

# ---------------- Data Structures ----------------

# Lists
numbers = [1, 2, 3, 4, 5]
numbers.append(6)
print(numbers)

# Dictionaries
data = {"name": "Alice", "age": 25, "city": "New York"}
print(data["name"])

# Tuples
coordinates = (10.5, 20.3)
print(f"Coordinates: {coordinates}")

# Sets
unique_numbers = {1, 2, 3, 4, 4, 5}
print(unique_numbers)

# Stacks (LIFO)
stack = []
stack.append(10)
stack.append(20)
print(stack.pop())

# Queues (FIFO)
from collections import deque
queue = deque([1, 2, 3])
queue.append(4)
print(queue.popleft())

# ---------------- File Handling ----------------

# Writing to a file
with open("sample.txt", "w") as file:
    file.write("Hello, this is a sample file!")

# Reading from a file
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)

# JSON Handling
import json
data_json = json.dumps(data)
print(data_json)

# ---------------- Mathematics for Machine Learning ----------------

# Linear Algebra
import numpy as np
matrix_A = np.array([[1, 2], [3, 4]])
matrix_B = np.array([[5, 6], [7, 8]])
print("Matrix Multiplication:", np.dot(matrix_A, matrix_B))

# Calculus - Derivative Approximation
def derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

def function(x):
    return x**2 + 3*x + 2

print("Approximate Derivative at x=1:", derivative(function, 1))

# Probability and Statistics
samples = np.random.normal(0, 1, 1000)
print(f"Mean: {np.mean(samples)}, Std Dev: {np.std(samples)}")

# ---------------- NumPy Basics ----------------

arr = np.array([1, 2, 3, 4, 5])
print(f"NumPy Array: {arr}")

# Reshaping Arrays
matrix = arr.reshape(5, 1)
print(matrix)

# Statistical Functions
print(f"Mean: {np.mean(arr)}, Standard Deviation: {np.std(arr)}")

# ---------------- Pandas Basics ----------------

import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
print(df)

# Handling Missing Data
df.loc[1, "Age"] = None
df.fillna(df["Age"].mean(), inplace=True)
print(df)

# Reading and Writing CSV
df.to_csv("sample_data.csv", index=False)
df_read = pd.read_csv("sample_data.csv")
print(df_read.head())

# ---------------- Data Visualization ----------------

import matplotlib.pyplot as plt
import seaborn as sns

# Simple Line Plot
plt.plot([1, 2, 3, 4, 5], [2, 4, 6, 8, 10])
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Seaborn Histogram
sns.histplot(df["Age"], bins=5, kde=True)
plt.show()

# ---------------- Working with Datasets ----------------

# Load Titanic Dataset
df_titanic = sns.load_dataset("titanic")
print(df_titanic.head())

# Data Analysis
print(f"Survival Rate: {df_titanic['survived'].mean() * 100:.2f}%")

# Visualization
sns.countplot(x="class", data=df_titanic)
plt.title("Passenger Class Distribution")
plt.show()
```
